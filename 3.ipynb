### 二、列与行操作

1. 选取 score 和 gender 两列。
2. 选取索引 0、2 的行。
3. 筛选 score>80 的行。
4. 新增一列 grade：score≥90 为优秀，否则为良好。
5. 删除 gender 列。

```python
import pandas as pd

#定义字典
data={
    'name':['Tom','Jerry','Mike'],
    'score':[85,92,78],
    'gender':['男','女','男']
}

#用字典创建 DataFrame
df=pd.DataFrame(data)

#1.选取 score 和 gender 两列。
#df[]->从df中取出数据，多列需打包为列表['','',……]
print("\n（1）选取 score 和 gender 两列")
columns=df[['score','gender']]
print(columns)

#2.选取索引 0、2 的行。
#iloc 位置索引 df.iloc[行，列]，多行[ , , ,……]
print("\n（2）选取索引 0、2 的行")
rows=df.iloc[[0,2]]
print(rows)

#3.筛选 score>80 的行。
#(原数据)df[(筛选条件)df['score']>80]
print("\n（3）筛选 score>80 的行")
filtered=df[df['score']>80]
print(filtered)

#4.新增一列 grade：score≥90 为优秀，否则为良好。
#使用numpy的where函数实现条件判断
#np.where(条件，True,False)
print("\n（4）新增一列 grade：score≥90 为优秀，否则为良好")
import numpy as np
df['grade']=np.where(df['score']>=90,'优秀','良好')
print(df)

#5.删除 gender 列。
#axis=0->行；axis=1->列;inplace=True:直接修改原表
print("\n（5）删除 gender 列")
df.drop('gender',axis=1,inplace=True)
print(df)
```

### 三、缺失值

1. 创建含缺失值的 DataFrame：

   data = {"x":[1, None, 3], "y":[None, 5, 6]}

   检测缺失值并统计每列缺失数量。

2. 用 0 填充缺失值；删除含缺失值的行。

```python
import pandas as pd
data={
    "x":[1,None,3],
    "y":[None,5,6]
}

#用字典创建 DataFrame
print("\n 1.创建含缺失值的 DataFrame")
df=pd.DataFrame(data)
print(df)
#检测缺失值 isnull
print("\n检测缺失值")
print(df.isnull())
#sum:True->1;False->0,每列相加
print("\n统计每列缺失数量")
print(df.isnull().sum())

#2.用 0 填充缺失值 fillna；删除含缺失值的行 dropna。
print("\n用 0 填充缺失值")
df_fill=df.fillna(0)
print(df_fill)
print("\n删除含缺失值的行")
df_drop=df.dropna()
print(df_drop)
```

### 四、分组聚合

1. 按 gender 分组，求 score 的均值和最大值。
2. 按 gender 分组，统计每组人数。

```python
import pandas as pd

#定义字典
data={
    'name':['Tom','Jerry','Mike'],
    'score':[85,92,78],
    'gender':['男','女','男']
}

#用字典创建 DataFrame
df=pd.DataFrame(data)
print(df)

#按 gender 分组
print("\n求 score 的均值和最大值")
#对gender分组（groupby），对不同组的score分别计算
#agg：应用数学函数；mean：均值；max:最大值
group=df.groupby('gender')['score'].agg(['mean','max'])
print(group)
print('\n')
group1=df.groupby('gender')['score'].agg(['mean'])
print(group1)

#size->统计每一组的数量（含空值）
print("\n统计每组人数")
count=df.groupby('gender').size()
print(count)
```

### 五、排序与去重

1. 按 score 升序排序。
2. 给 DataFrame 添加一行重复数据（如 Tom,85, 男），去重。

```python
import pandas as pd

#定义字典
data={
    'name':['Tom','Jerry','Mike'],
    'score':[85,92,78],
    'gender':['男','女','男']
}

#用字典创建 DataFrame
df=pd.DataFrame(data)
print(df)

#sort_values(排序)，by=……(排序依据)，ascending(True升序/从小到大,False降序)
print("\n按 score 升序排序")
sort=df.sort_values(by='score',ascending=True)
print(sort)

#给 DataFrame 添加一行重复数据（如 Tom,85, 男），去重。
#{……}字典，[]列表，包裹字典，此时字典成为列表的一个元素
new=pd.DataFrame([{
    'name':'Tom',
    'score':85,
    'gender':'男'
}])#此时构造了一个DataFrame：new，里有一行数据
#dup(重复);pd.concat[拼接];ignore_index=True(重排序号)
dup=pd.concat([df,new],ignore_index=True)
print("\n添加重复数据")
print(dup)

print("\n去重")
clean=dup.drop_duplicates()
print(clean)
```

### 六、合并与拼接

1. 横向合并：

   df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})

   df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})

   按 id 合并。

```python
import pandas as pd
#pd.DataFrame(调用函数创建表格)；字典：{"键":[值]}
df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})
df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})

#pd.merge 合并函数；on='……'连接键
print("按ID横向合并：")
merge=pd.merge(df1,df2,on='id')
print(merge)
```

2. 纵向拼接：

   df_a = pd.DataFrame ({"val":[1,2]})

   df_b = pd.DataFrame ({"val":[3,4]})

   拼接后重置索引。

```python
df_a=pd.DataFrame({"val":[1,2]})
df_b=pd.DataFrame({"val":[3,4]})
print("纵向拼接后重置索引")
#concat 拼接 ignore_index=Ture 重置索引
concat=pd.concat([df_a,df_b],ignore_index=True)
print(concat)
```

### 七、时间处理

1. 创建时间列：["2025-01-01","2025-01-02","2025-01-03"]，转为 datetime 格式。
2. 提取时间列的月份、星期。

```python
date=["2025-01-01","2025-01-02","2025-01-03"]
#pd.to_datetime() 把字符串列表转换为datetime格式
datetime=pd.to_datetime(date)
#创建DataFrame表格
df_time=pd.DataFrame({'date':datetime})
print("时间列：")
print(df_time)

print("\n现数据类型：")
print(df_time['date'].dtype)

print("\n提取时间列的月份、星期:")
#.dt->时间属性的访问器；从df_time['date']里提取month
#填入df_time表格的新列 month
#weekday->0=Monday~6=Sunday
#day_name->Monday~Sunday
df_time['month']=df_time['date'].dt.month
df_time['weekday']=df_time['date'].dt.weekday
df_time['day']=df_time['date'].dt.day_name()
print(df_time)
```

### 八、高级操作

1. 对 score 列做标准化（(值 - 均值)/ 标准差）。
2. 用透视表：行是 gender，值是 score，聚合函数为中位数。

```python
import pandas as pd

#定义字典
data={
    'name':['Tom','Jerry','Mike'],
    'score':[85,92,78],
    'gender':['男','女','男']
}

#1.用字典创建 DataFrame
df=pd.DataFrame(data)
print(df)

#从df中拿出score->值
scores=df['score']

print("\n均值：")
mean_score=scores.mean()
print(mean_score)

print("\n标准差：")
std_score=scores.std()
print(std_score)

print("\n对 score 列做标准化（(值 - 均值)/ 标准差）:")
df['score_normalized']=(scores-mean_score)/std_score
print(df)

#用透视表 pivot table
#行是 gender，值是 score
#聚合函数为中位数
#创建透视表：pd.pivot_table(数据源df,值score,索引/行gender,聚合统计函数aggfunc)
#aggfunc->默认是均值mean，若有修改需要写 “='……'”
pivot_table=pd.pivot_table(df,values='score',index='gender',aggfunc='median')
print("\n透视表：")
print(pivot_table)
```
